## Problem 1: Social Media Relationships (9 points) 

Let us pretend that you are building a new app. Your app supports social media functionality, which means that users can have _relationships_ with other users.  

There are two guidelines for describing relationships on this social media app:  
1. Any user can _follow_ any other user.  
2. If two users follow each other, they are considered _friends_.  

Data about your users and their followers is stored in a dictionary that adheres to this structure:  

```python
# NON-INTERACTIVE CODE CELL. YOU MAY RUN THIS CELL, BUT DO NOT EDIT IT.
# FOR DEMONSTRATION PURPOSES ONLY. DO NOT EDIT.

social_graph = {
    "@bongolpoc":{"first_name":"Joselito",
                  "last_name":"Olpoc",
                  "following":[
                  ]
    },
    "@joaquin":  {"first_name":"Joaquin",
                  "last_name":"Gonzales",
                  "following":[
                      "@chums","@jobenilagan"
                  ]
    },
    "@chums" : {"first_name":"Matthew",
                "last_name":"Uy",
                "following":[
                    "@bongolpoc","@miketan","@rudyang","@joeilagan"
                ]
    },
    "@jobenilagan":{"first_name":"Joben",
                   "last_name":"Ilagan",
                   "following":[
                    "@eeebeee","@joeilagan","@chums","@joaquin"
                   ]
    },
    "@joeilagan":{"first_name":"Joe",
                  "last_name":"Ilagan",
                  "following":[
                    "@eeebeee","@jobenilagan","@chums"
                  ]
    },
    "@eeebeee":  {"first_name":"Elizabeth",
                  "last_name":"Ilagan",
                  "following":[
                    "@jobenilagan","@joeilagan"
                  ]
    },
}
```

**Write a function called `relationship_status` that takes three positional arguments: (str) `from_member`, (str) `to_member`, and (dict) `social_graph`. The function should determine the _relationship status_ of the first member to the second member based on the data stored in the social graph. The function should _return_ one of these values depending on what is appropriate:**  
- "follower", if `from_member` follows `to_member`
- "followed by", if `from_member` is followed by `to_member`
- "friends", if `from_member` and `to_member` follow each other
- None if none of the above scenarios are applicable  

Specifications:  
1. `from_member` and `to_member` are user IDs (e.g. "@jobenilagan").
2. `social_graph` is a dictionary that adheres to the specification demonstrated in the demo cell.

```python
# CODE CELL

def relationship_status(from_member,to_member,social_graph):
    try:
        if False == all([p != to_member for p in social_graph[from_member]['following']]) and False == all([p != from_member for p in social_graph[to_member]['following']]):
            return "friends" #friends
        elif False == all([p != to_member for p in social_graph[from_member]['following']]): 
            return "follower" #follower
        elif False == all([p != from_member for p in social_graph[to_member]['following']]):
            return f"followed by:{to_member}" #followed by
        else:
            "None"
    except: 
        return "Invalid Input"
        
        
# PROBLEM 1
print(relationship_status('@joeilagan','@chums',social_graph))
print(relationship_status('@chums','@joeilagan',social_graph))

print(relationship_status('@bongolpoc','@joaquin',social_graph))
print(relationship_status('@joaquin','@bongolpoc',social_graph))

print(relationship_status('@chums','@bongolpoc',social_graph))
print(relationship_status('@bongolpoc','@chums',social_graph))

print(relationship_status('chums','@joeilagan',social_graph))
print(relationship_status('@chums','joeilagan',social_graph))

print(relationship_status(1,'@joeilagan',social_graph))
print(relationship_status('@chums',False,social_graph))
```

## Problem 2: Tic Tac Toe (10 points)  

Tic Tac Toe is a common paper-and-pencil game. Players must attempt to successfully draw a straight line of their symbol across a grid. The player that does this first is considered the winner.  

Here are several possible board configurations:

```python
# NON-INTERACTIVE CODE CELL. YOU MAY RUN THIS CELL, BUT DO NOT EDIT IT.
# FOR DEMONSTRATION PURPOSES ONLY. DO NOT EDIT.

board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]
```

**Write a function called `tic_tac_toe` that takes one positional argument (list) `board`. The function should _return_ the winner of the board if there is one, and it should return `None` if there is no winner on the board.**  

Specifications:
1. Each player is represented by their symbol. Example: if the player using 'X' won the board, then simply return the string 'X'.
2. The board may be 3x3, 4x4, 5x5, or 6x6.
3. The game will only ever be player X vs player O. No other symbols will be used.
4. There may be no winner, and there may be 1 winner, but there will never be a situation where there is more than 1 winner.

```python
# CODE CELL

FourxFourBoard = [
    ['O','X','X','X'],
    ['X','O','O','X'],
    ['O','X','O','X'],
    ['X','X','O','O'],
] 

FivexFiveBoard = [
    ['X','X','X','X','X'],
    ['X','O','X','O','X'],
    ['O','X','O','X','X'],
    ['X','X','O','X','O'],
    ['X','O','X','X','X']
]

SixxSixBoard = [
    ['X','O','X','X','X','X'],
    ['X','O','X','O','X','X'],
    ['X','X','O','X','X','O'],
    ['X','X','O','X','O','X'],
    ['X','O','X','X','X','O'],
    ['X','O','X','X','X','X']
]

def tic_tac_toe(board):
    try:
    #Winner
        #Diagonal Up-Down
        if False != all([board[p][p] == 'X' for p in range(len(board))]):
            return "X Wins!"

        elif False != all([board[p][p] == 'O' for p in range(len(board))]):
            return "O Wins!"

        #Diagonal Up-Down
        elif False != all([board[2-p][p] == 'X' for p in range(len(board))]):
            return "X Wins!"

        elif False != all([board[2-p][p] == 'O' for p in range(len(board))]):
            return "O Wins!"

        #Horizontal
        elif False == all([board[p] != ['X','X','X'] for p in range(len(board))]):
            return "X Wins!"
        elif False == all([board[p] != ['X','X','X','X'] for p in range(len(board))]):
            return "X Wins!"
        elif False == all([board[p] != ['X','X','X','X','X'] for p in range(len(board))]):
            return "X Wins!"
        elif False == all([board[p] != ['X','X','X','X','X','X'] for p in range(len(board))]):
            return "X Wins!"
        
        elif False == all([board[p] != ['O','O','O'] for p in range(len(board))]):
            return "O Wins!"
        elif False == all([board[p] != ['O','O','O','O'] for p in range(len(board))]):
            return "O Wins!"
        elif False == all([board[p] != ['O','O','O','O','O'] for p in range(len(board))]):
            return "O Wins!"
        elif False == all([board[p] != ['O','O','O','O','O','O'] for p in range(len(board))]):
            return "O Wins!"

        #Vertical
        elif False == all([p != ('X','X','X') for p in zip(*board)]):
            return "X Wins!"
        elif False == all([p != ('X','X','X','X') for p in zip(*board)]):
            return "X Wins!"
        elif False == all([p != ('X','X','X','X','X') for p in zip(*board)]):
            return "X Wins!"
        elif False == all([p != ('X','X','X','X','X','X') for p in zip(*board)]):
            return "X Wins!"
       
        elif False == all([p != ('O','O','O') for p in zip(*board)]):
            return "O Wins!"
        elif False == all([p != ('O','O','O','O') for p in zip(*board)]):
            return "O Wins!"
        elif False == all([p != ('O','O','O','O','O') for p in zip(*board)]):
            return "O Wins!"
        elif False == all([p != ('O','O','O','O','O','O') for p in zip(*board)]):
            return "O Wins!"

    #No Winner
        #Diagonal Up-Down
        elif False == all([board[p][p] == 'X' for p in range(len(board))]):
            return "No Winner!"

        elif False == all([board[p][p] == 'O' for p in range(len(board))]):
            return "No Winner!"

        #Diagonal Down-Up
        elif False == all([board[2-p][p] == 'X' for p in range(len(board))]):
            return "No Winner!"

        elif False == all([board[2-p][p] == 'O' for p in range(len(board))]):
            return "No Winner!"

        #Horizontal
        elif False != all([board[p] != ['X','X','X'] for p in range(len(board))]):
            return "No Winner!"
        elif False != all([board[p] != ['X','X','X','X'] for p in range(len(board))]):
            return "No Winner!"
        elif False != all([board[p] != ['X','X','X','X','X'] for p in range(len(board))]):
            return "No Winner!"
        elif False != all([board[p] != ['X','X','X','X','X','X'] for p in range(len(board))]):
            return "No Winner!"

        elif False != all([board[p] != ['O','O','O'] for p in range(len(board))]):
            return "No Winner!"
        elif False != all([board[p] != ['O','O','O','O'] for p in range(len(board))]):
            return "No Winner!"
        elif False != all([board[p] != ['O','O','O','O','O'] for p in range(len(board))]):
            return "No Winner!"
        elif False != all([board[p] != ['O','O','O','O','O','O'] for p in range(len(board))]):
            return "No Winner!"


        #Vertical
        elif False != all([p != ('X','X','X') for p in zip(*board)]):
            return "No Winner!"
        elif False != all([p != ('X','X','X','X') for p in zip(*board)]):
            return "No Winner!"
        elif False != all([p != ('X','X','X','X','X') for p in zip(*board)]):
            return "No Winner!"
        elif False != all([p != ('X','X','X','X','X','X') for p in zip(*board)]):
            return "No Winner!"
        
        elif False != all([p != ('O','O','O') for p in zip(*board)]):
            return "No Winner!"
        elif False != all([p != ('O','O','O','O') for p in zip(*board)]):
            return "No Winner!"
        elif False != all([p != ('O','O','O','O','O') for p in zip(*board)]):
            return "No Winner!"
        elif False != all([p != ('O','O','O','O','O','O') for p in zip(*board)]):
            return "No Winner!"
        
        else:
            return "Invalid Input"
    except:
        return("Invalid Input")


# PROBLEM 2
print(tic_tac_toe(board1))
print(tic_tac_toe(board2))
print(tic_tac_toe(board3))
print(tic_tac_toe(board4))
print(tic_tac_toe(board5))
print(tic_tac_toe(board6))
print(tic_tac_toe(FourxFourBoard))
print(tic_tac_toe(FivexFiveBoard))
print(tic_tac_toe(SixxSixBoard))
print(tic_tac_toe(1))
print(tic_tac_toe(True))
```

## Problem 3: Routing (9 points)

During the pandemic, a shuttle van service is tasked to travel along a predefined circular route as follows:

- UP Diliman -> Ateneo de Manila (Estimated Time: 10 mins)
- Ateneo de Manila -> De La Salle Taft (Estimated Time: 35 mins)
- De La Salle Taft -> UP Diliman (Estimated Time: 55 mins)

The route is one-way only. So, the van cannot go back directly to UP Diliman from Ateneo de Manila; rather, it will have to drop by De La Salle Taft first before going to UP Diliman.



```python
# NON-INTERACTIVE CODE CELL. YOU MAY RUN THIS CELL, BUT DO NOT EDIT IT.
# FOR DEMONSTRATION PURPOSES ONLY. DO NOT EDIT.

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}
```

**Write a function called `eta` that takes three positional arguments: (dict) `legs`, (str) `source`, and (str) `destination`. The function should _return_ the estimated number of minutes it will take to reach the destination from the source based on the data stored in the `legs` dictionary.**  

Specifications:
1. The `legs` dictionary adheres to the schema followed in the demo cell.  
2. There may be more legs than shown in the demo cell.
3. Any destination can be reached from any source.

```python
# CODE CELL

upd_admu = legs[('upd','admu')]['travel_time_mins']
admu_dlsu = legs[('admu','dlsu')]['travel_time_mins']
dlsu_upd = legs[('dlsu','upd')]['travel_time_mins']


def eta(legs,source,destination):
    try:
        #from upd
        if source == 'upd' and destination == 'admu':
            return str(upd_admu) + " minutes"
        elif source == 'upd' and destination == 'dlsu':
            return str(upd_admu + admu_dlsu) + " minutes"
        elif source == 'upd' and destination == 'upd':
            return str(0) + " minutes"

        #from admu
        elif source == 'admu' and destination == 'dlsu':
            return str(admu_dlsu) + " minutes"
        elif source == 'admu' and destination == 'upd':
            return str(admu_dlsu + dlsu_upd) + " minutes"
        elif source == 'admu' and destination == 'admu':
            return str(0) + " minutes"

        #from dlsu
        elif source == 'dlsu' and destination == 'upd':
            return str(dlsu_upd) + " minutes"
        elif source == 'dlsu' and destination == 'admu':
            return str(dlsu_upd + upd_admu) + " minutes"
        elif source == 'dlsu' and destination == 'dlsu':
            return str(0) + " minutes"

        else: 
            return "A location is not available or invalid"
    except:
        pass


# PROBLEM 3
print(eta(legs,'upd','admu'))
print(eta(legs,'upd','dlsu'))
print(eta(legs,'admu','dlsu'))
print(eta(legs,'admu','upd'))
print(eta(legs,'dlsu','upd'))
print(eta(legs,'dlsu','admu'))
print(eta(legs,'dlsu','dlsu'))
print(eta(legs,'upd','upd'))
print(eta(legs,'admu','admu'))
print(eta(legs,'feu','dlsu'))
print(eta(legs,1,'dlsu'))
print(eta(legs,1,True))
```
